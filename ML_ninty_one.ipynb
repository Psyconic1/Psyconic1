```python
path=r'C:\Users\DELL\Desktop\tes\final'
dstpath=r'C:\Users\DELL\Desktop\testm'
files=list(filter(lambda f: isfile(join(path,f)),listdir(path)))
for image in files:
    try:
        img=cv2.imread(os.path.join(path,image))
        cubic_img = cv2.resize(img,(224, 224), fx = 2, fy = 2, interpolation = cv2.INTER_CUBIC)
        median=cv2.medianBlur(cubic_img,3)
        gauss=cv2.GaussianBlur(median,(3,3),0)
        twoDim = gauss.reshape((-1,3))
        twoDim = np.float32(twoDim)
        criteria = (cv2.TERM_CRITERIA_EPS + cv2.TERM_CRITERIA_MAX_ITER , 10 , 1.0)
        K=4
        attamps = 15
        ret , label , center = cv2.kmeans(twoDim, K, None , criteria , attamps, cv2.KMEANS_PP_CENTERS)
        center = np.uint8(center)
        res = center[label.flatten()]
        result_image = res.reshape((gauss.shape))
        dstPath = join(dstpath,image)
        cv2.imwrite(dstPath, result_image)
    except:
        print("aaa")
```

```python
start = timeit.default_timer() 

df = pd.read_csv(r'C:\Users\DELL\Desktop\testm\hogf.csv')
dataset=df.astype(float)
X = dataset.drop('Target',axis=1)
y = dataset['Target']

x_train, x_test, y_train, y_test = train_test_split(X, y, test_size = 0.25, stratify = y, random_state=0)


# MODEL -> SVM CLASSIFICATION (kernel : rbf)
classification_rbf = SVC(kernel="poly",C=9, degree=3, gamma=0.1, coef0=0.0, shrinking=False, probability=True, tol=0.001, cache_size=100, class_weight=None, verbose=False, max_iter=-1, decision_function_shape='ovr', random_state=4)
hist=classification_rbf.fit(x_train, y_train)
predicts_rbf = classification_rbf.predict(x_test)
df.head()
```

```python
def plot_confusion_matrix(cm, classes,normalize=True,
                          title='Confusion matrix',
                          cmap=plt.cm.Blues):
    if normalize:
        cm = cm.astype('float') / cm.sum(axis=1)[:, np.newaxis]
    plt.imshow(cm, interpolation='nearest', cmap=cmap,origin='upper')
    plt.title(title)
    tick_marks = np.arange(len(classes))
    plt.xticks(tick_marks, classes, rotation=70)
    plt.yticks(tick_marks, classes)
    fmt = '.2f' 
    thresh = cm.max() / 1.5 if normalize else cm.max() /2
    thresh = cm.max() / 2.
    for i, j in itertools.product(range(cm.shape[0]), range(cm.shape[1])):
        plt.text(j, i, format(cm[i, j], fmt),
                 horizontalalignment="center",
                 color="white" if cm[i, j] > thresh else "black")
    plt.tight_layout()
    plt.ylabel('True label')
    plt.xlabel('Predicted label')
    plt.grid(False)
    plt.show()
target_names = ['bacterial cancer','cutting weevel','Die back','Gull midge','Healthy']

cm_rbf = confusion_matrix(y_test, predicts_rbf)
print("Confusion Matrix (RBF Kernel):\n"+str(cm_rbf)) #radial basis function kernel
plot_confusion_matrix(cm_rbf, target_names)

#print("Accuracy: "+str( (cm_rbf[0][0]+cm_rbf[1][1]) / (sum(cm_rbf[0])+sum(cm_rbf[1])) ) )
print("Test Accuracy of SVM Algorithm (RBF): {:.2f}%".format(classification_rbf.score(x_test,y_test)*100))
print("Train Accuracy of SVM Algorithm (RBF): {:.2f}%".format(classification_rbf.score(x_train,y_train)*100))
print("TPR rate (Sensitivity-Recall) : "+ str(cm_rbf[1][1]) +" / "+str(cm_rbf[1][0]+cm_rbf[1][1])+" = "+str(cm_rbf[1][1]/(cm_rbf[1][0]+cm_rbf[1][1]) )  ) 

from sklearn.metrics import classification_report
print(classification_report((y_test), predicts_rbf,target_names=target_names))
    
    
        
    
```
